```python
# column NAme
column_name = ['sales_number','sales_order_line','Order_date','customer_Name','Email_Addree','Item','Quantity','unit_Price','Taxt_Amount']
```

```python
# read the csv file with column name to use of pandas
df = pd.read_csv('2021.csv',names=column_name)
```

```python
# duplicate row

duplicate = df[df.duplicated()]
if not duplicate.empty:
    print("duplicate" , duplicate)
else:
    print("No duplicate rows")
```

```python
# split the Order_date column into year and month 
df['day', 'Month'] = df['Order_date'].str.slice(8, 10 )
df
```

```python
order_date_index= df.columns.get_loc('Order_date')

# insert year column after order_date

df.insert(order_date_index + 1 'year' , df.pop('year'))
df.insert(order_date_index + 1 'month' , df.pop('month'))
```

```python
# drop the day column

df.drop('day', axis=1, inplace=True)
```
